```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sads.csv")
df
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
stop = stopwords.words('english')
test = df
```

```python
X = df.review
y = df.sentiment
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
Xtrain_vects = vectorizer.fit_transform(X_train)
Xtest_vects = vectorizer.transform(X_test)
```

## MODEL


```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0).fit(Xtrain_vects, y_train)
```

```python
score = clf.score(Xtest_vects, y_test)
print(score*100, '%')
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, clf.predict(Xtest_vects), average = None)*100
```

```python
test = ['Loved the acting of every character in it. Reality was shown with lots of information in it. Sprinkling some of the beautifully composed music and sung by the great singers', 'this movie was not good at all', 'movie was not good']
pred_val = vectorizer.transform(test)
clf.predict(pred_val)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sads.csv")
sentiment = df['sentiment'].replace({'positive': 0, 'negative': 1})
df['sentiment'] = sentiment
df.head(10000)
df
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
stop = stopwords.words('english')
test = df
```

```python
X = df.review
y = df.sentiment
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
Xtrain_vects = vectorizer.fit_transform(X_train)
Xtest_vects = vectorizer.transform(X_test)
```

## MODEL AND TUNING

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear')
clf.fit(Xtrain_vects, y_train)
```

```python
clf.score(Xtest_vects, y_test)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, clf.predict(Xtest_vects), average = None)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
url = "https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv"
df = pd.read_csv(url)
df.head(15)
```

```python
df.rename(columns={'per capita income (US$)': 'PCI'}, inplace=True)
df.head(15)
```

```python
X = df.year
y = df.PCI
plt.plot(X, y)
plt.title('Year')
plt.xlabel('Per Capita Income in Canada(1970-2016)')
plt.ylabel('PerCapitaIncome(USD)')
plt.show()
```

```python
X = df.year
y = df.PCI
from sklearn.linear_model import LinearRegression 
reg = LinearRegression()
reg.fit(df[['year']], df[['PCI']])
prediction = reg.predict(df[['PCI']])
```
